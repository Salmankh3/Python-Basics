```python
import math
import sys
current_list=[]
resistance_list=[]
voltage_list=[]
power_list=[]
calculation= int(input("How many calculations you want to perform?"))
for i in range(calculation):
    print("Select Operation:")
    print("1. To calculate power and resistance")
    print("2. To calculate power and voltage")
    print("3. To calculate power and current")
    print("4. To calculate voltage and resistance")
    print("5. To calculate current and resistance")
    print("6. To calculate voltage and current")
    operation=int(input())
    if operation==1:
     print("Enter the value of current")
     current=float(input())
     current_list.append(current)
     print("Enter the value of voltage")
     voltage=float(input())
     voltage_list.append(voltage)
     power = voltage * current
     power_list.append(power)
     resistance = voltage / current
     resistance_list.append(resistance)
     print("power="+str(power))
     print("resistance="+str(resistance))
    elif operation==2:
     print("Enter the value of resistance:")
     resistance=float(input())
     resistance_list.append(resistance)
     print("Enter the value of current")
     current=float(input())
     current_list.append(current)
     power = current * current * resistance
     power_list.append(power)
     voltage = current * resistance
     print("power= "+str(power))
     print("voltage= "+str(voltage))
    elif operation==3:
     print("Enter the value of voltage:")
     voltage=float(input())
     voltage_list.append(voltage)
     print("Enter the value of resistance:")
     resistance=float(input())
     resistance_list.append(resistance)
     power=(voltage * voltage)/resistance
     power_list.append(power)
     current=voltage/resistance
     current_list.append(current)
     print("power= "+str(power))
     print("current= "+str(current))
    elif operation==4:
     print("Enter the value of current:")
     current=float(input())
     current_list.append(current)
     print("Enter the value of power:")
     power=float(input())
     power_list.append(power)
     resistance = power / (current * current)
     resistance_list.append(resistance)
     voltage= current * resistance
     voltage_list.append(voltage)
     print("resistance= "+str(resistance))
     print("voltage= "+str(voltage))
    elif operation==5:
     print("Enter the value of voltage:")
     voltage=float(input())
     voltage_list.append(voltage)
     print("Enter the value of power:")
     power=float(input())
     power_list.append(power)
     current = power / voltage
     current_list.append(current)
     resistance = (voltage * voltage) / power
     print("current= "+str(current))
     print("resistance= "+str(resistance))
    elif operation==6:
     print("Enter the value of Resistance:")
     resistance=float(input())
     resistance_list.append(resistance)
     print("Enter the value of power:")
     power=float(input())
     power_list.append(power)
     voltage = math.sqrt(power * resistance)
     voltage_list.append(voltage)
     current = voltage /resistance
     current_list.append(current)
     print("voltage= "+str(voltage))
     print("current= "+str(current))
    else:
     sys.exit()
```
